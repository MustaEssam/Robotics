```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
# Uncomment the next line for use in a Jupyter notebook
# This enables the interactive matplotlib window
%matplotlib notebook
image = mpimg.imread('example_grid1.jpg')
plt.imshow(image)
plt.show() 
```

```python
import cv2
import numpy as np

def perspect_transform(img, src, dst):

    # Get transform matrix using cv2.getPerspectivTransform()
    M = cv2.getPerspectiveTransform(src, dst)
    # Warp image using cv2.warpPerspective()
    # keep same size as input image
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    # Return the result
    return warped

x_size, y_size = image.shape[1::-1]
# Define source and destination points
source = np.float32([[14 ,140], [118 ,96], [200 ,96], [301 ,140]])
#square size will be 10x10 pixels 
destination = np.float32([[x_size/2-5, y_size-5], [x_size/2-5, y_size-15],
                          [x_size/2+5, y_size-15], [x_size/2+5, y_size-5]])      

warped = perspect_transform(image, source, destination)
plt.figure()
plt.imshow(warped)
plt.show() 
```
